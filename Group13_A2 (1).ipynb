```python
import pandas as pd

def file_meld_pd(input_name_1, input_name_2, output_name):
    input_name_1 = pd.read_csv("A2_input1.csv", index_col=False)  #using pandas we read the content of the first file
    input_name_2 = pd.read_csv("A2_input2.csv", index_col=False)  #using pandas we read the content of the second file
    output_name=input_name_1.merge(input_name_2, on ="id")  #using pandas we merge both files
    
    return output_name

    
    # YOUR CODE HERE
```
